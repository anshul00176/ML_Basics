```python
#now we will make our GD class for performing GD
class GDRegressor:
    #we will take learning rate ,no of epochs, intercept, m from the user only 
    def __init__ (self,learning_rate,epochs,intercept,m):
      self.learning_rate=learning_rate
      self.m=m
      self.epochs=epochs
      self.intercept=intercept

    def fit(self,X,y):

      #now we try to mimimize value of b and m epochs no of time

      for i in range(self.epochs):
        #calculating del(l)/del(b)  and m is already taken as constant
        loss_slope_b=-2*np.sum(y-self.m*X.ravel()-self.intercept)
        loss_slope_m=-2*np.sum( (y-self.m*X.ravel()-self.intercept)*X.ravel())

         #now caluclate new intercept
        self.intercept=self.intercept-(self.learning_rate*loss_slope_b)
        self.m=self.m-(self.learning_rate*loss_slope_m)

        print(self.intercept, self.m)
        print()


```

```python
gd = GDRegressor(0.1,10,-120,147)
```

```python
gd.fit(X,y)
# look at  the intercept and slope value how it is shifting from -ve to poisitve
# we need value near 0.91( from actual sklearn LR)
#we are now making chages into learning rate and do the training again
```
